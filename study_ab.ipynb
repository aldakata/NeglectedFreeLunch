```python
def load_points(xml_file):
    tree = ET.parse(xml_file)
    root = tree.getroot()

    points = []
    for obj in root.findall("metadata"):
        if obj.find("selected").text == "True":
            selected_point = obj.find("selectedRecord")
            points.append(float(selected_point.find("x").text))
            points.append(float(selected_point.find("y").text))
            points.append(float(selected_point.find("time").text))
            selected_point = obj.find("hoveredRecord")
            points.append(selected_point.find("action").text)
            points.append(float(selected_point.find("time").text))
    # print(points)
    return np.array(points)
```

```python
def load_points(xml_file):
    tree = ET.parse(xml_file)
    root = tree.getroot()
    estimate_time = 0
    points = []
    for obj in root.findall("metadata"):
        if obj.find("selected").text == "True":
            estimate_time = int(obj.find("estimateTime").text)
            
            selected_point = obj.find("selectedRecord")
            points.append(
                [
                    float(selected_point.find("x").text),
                    float(selected_point.find("y").text),
                ]
            )

    return np.array(points), estimate_time
```
